Based on hiring statics for a firm such as experience of candidate, his written test score and personal interview score, HR will decide the salary. Given this data, build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates:

a. 2 yr experience, 9 test score, 6 interview score
b. 12 yr experience, 10 test score, 10 interview score

**The model is built on Scikit Learn Multivariate Linear Regression**

```python
!pip install word2number
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/aaaraafaat/Data-Science/refs/heads/main/hiring.csv")
df
```

```python
df.experience = df.experience.fillna("zero")
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
```

```python
df
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())
```

```python
df
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(int(df['test_score(out of 10)'].mean()))
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('salary($)', axis='columns'), df['salary($)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```
